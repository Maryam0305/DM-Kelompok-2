```python
import pandas as pd

# Load dataset
data = pd.read_csv('data/dataSet2.csv')
data.head()
```

```python
data.info()
print(data.isnull().sum())
```

```python
# Contoh membersihkan data (sesuaikan dengan kebutuhan dataset Anda)
data = data.dropna()
data.to_csv('data/cleaned_dataSet2.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribusi data
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
```
